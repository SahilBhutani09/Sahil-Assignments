```python
#XGES Causal discovery algorithm implementation in Python
#It stands for "eXtended Greedy Equivalence Search"
#XGES is an extension of the GES algorithm that allows for more flexible search strategies in causal discovery.
#XGES is designed to identify causal structures from observational data by searching through equivalence classes of directed acyclic graphs (DAGs).
#Implementation - It starts with an empty graph and iteratively adds, removes, or reverses edges based on a scoring criterion until a local optimum is reached.
```

```python
data = pd.read_csv("C:/Users/sahil/Downloads/archive/Student_Performance.csv")
```

```python
test = data.copy()
```

```python
test = test.drop(['Extracurricular Activities'], axis = 1)
def standardise(x):
    return (x -x.mean())/(x.max() - x.min())
test = test.apply(standardise, axis = 1)
test.corr()
```

```python
import random


def introduce_noise(x):
    return x + random.randint(0,10)/10

test = test.apply(introduce_noise, axis = 0)
```

```python
test_np = test.to_numpy()
test_np
```

```python
# run FCI
X = test_np
xges = XGES()
print(f"Running XGES on data with shape {X.shape} ...")
result = xges.fit(X = X)
```

```python
# Suppose 'pdag' is your XGES PDAG object
adj = result.to_adjacency_matrix()  # numpy array, shape (n_vars, n_vars)
type(adj)
```

```python
for i in range(len(adj)):
    for j in range(len(adj[i])):
        if adj[i][j] == 1:
            print(list(test)[i], '->', list(test)[j])
```
