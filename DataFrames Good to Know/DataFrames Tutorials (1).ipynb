```python
#Group by works with aggregations and transformations
#Aggregation produces a new data frame with shape of the results
#Transaformation retains the shape of original data frame and only works on one columns
#Aggregation functions - sum, mean, min, max, median, count(without na), size(with na)
#nunique, first last
```

```python
# How to normalise data with transforms

# Min-Max Normalization
def min_max_scaler(x):
    return (x - x.min()) / (x.max() - x.min())

def z_score_scaler(x):
    return (x - x.mean()) / x.std()

def log_scaler(x):
    return np.log(x)

df["NormalisedContractRenewalByAccount"] = df.groupby("AccountWeeks")["ContractRenewal"].transform(min_max_scaler)
```

```python
indx = input()
people = []
for i in range(0, int(indx)):
    people.append(input())
```

```python
df = pd.DataFrame({
    "first_name": [x.split(' ')[0] for x in people],
    "last_name": [x.split(' ')[1] for x in people],
    "age": [x.split(' ')[2] for x in people],
    "sex": [x.split(' ')[3] for x in people]
})
```

```python
df["age"] = df["age"].astype(int)
df.sort_values(["age"], ascending = [True])
```

```python
df["salu"] = ''
for i in range(len(df["sex"])):
    if df["sex"][i] == "M":
        df["salu"][i] = "Mr."
    else:
        df["salu"][i] = "Mrs."
```
