```python
submission = {
    "submissionID":[1,2,3,4],
    "submissionType": ['submission','article_transfer','submission','submission'],
    "authorId" : ['1','1','3','4'], 
    "decisionId" : ['2','1','2','3']
}
decision = {
    "decisionId" : ['1','2','3','4'], 
    "editorDecision" : ['Accept','Reject','Reject','Reject'],
    "decisionReason" : ['null','wrong scope','null','wrong scope']
}
author = {
    "authorId" : ['1','2','3','4'], 
    "authorName" : ['Maria','Nushi','Mohammed','Jose'], 
    "university" : ['University A','University B','University A','University C']
}
submissionDf = pd.DataFrame(submission)
decisionDf = pd.DataFrame(decision)
authorDf = pd.DataFrame(author)
```

```python
df = submissionDf.copy()
df = df.merge(decisionDf, on = "decisionId", how = "left").merge(authorDf, on = "authorId", how = "left")
tempDf = df.loc[df["decisionReason"] == "wrong scope"]
tempDf["index"] = 1
tempDf = tempDf.pivot_table(index = ["authorName", "university"], values = ["index"], aggfunc = sum).sort_values("index")
tempDf.iloc[0,]
```
