```python
#Base conversion Rate
Conv_rate = input("Enter base conversion rate in %")
uplift = input("Uplift that is expected from this test in %")
Power = input("Power in %")
Confidence_level = input("Confidence level in %")
```

```python
Conv_rate = float(Conv_rate.replace("%", ""))/100
uplift = float(uplift.replace("%", ""))/100
Power = float(Power.replace("%", ""))/100
Confidence_level = float(Confidence_level.replace("%", ""))/100
```

```python
uplift = Conv_rate * (1 + uplift) - Conv_rate
```

```python
visitors = input("Enter number of visitors per day")
visitors = int(visitors)
```

```python
import math
import numpy as np
import scipy.stats as stats
```

```python

variance = Conv_rate * (1 - Conv_rate)
alpha = 1 - Confidence_level
z_significance = abs(stats.norm.ppf (1 - alpha/2))
z_power = abs(stats.norm.ppf(1 - Power))

number_of_visitors = (2 * variance * math.pow((z_significance + z_power), 2)) / math.pow(uplift, 2)
```

```python
number_of_days = number_of_visitors / visitors
```

```python
print("Number of Days that we need to run this test is {number_of_days}".format(number_of_days = number_of_days))
```
