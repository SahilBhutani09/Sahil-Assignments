```python
## To Test if the test data set is significantly diffferent from control data set
```

```python
## Discreate Data Sets
```

```python
Visitors_A = input("Enter number of visitors for Control Data Set")
Visitors_B = input("Enter number of visitors for Test Data Set")

Conv_A = input("Enter number of conversions for Control Data Set")
Conv_B = input("Enter number of conversions for Test Data Set")
```

```python
Null_Hypthesis = "There is no significant deviation between test and control"
Alternate_Hypothesis = 'Test is significantly better than control'
```

```python
Conv_A  = float(Conv_A)
Conv_B = float(Conv_B)
Visitors_A  = float(Visitors_A)
Visitors_B = float(Visitors_B)

Conv_rate_A = Conv_A / Visitors_A
Conv_rate_B = Conv_B / Visitors_B
```

```python
confidence_level = 0.95 #We can change it for different tests
```

```python
import math
Z_Score = (Conv_rate_A - Conv_rate_B) / math.sqrt(
        (Conv_rate_A * (1 - Conv_rate_A) / Visitors_A)
        + 
        (Conv_rate_B * (1 - Conv_rate_B) / Visitors_B)
)
```

```python
import numpy as np
import scipy.stats as stats

# Calculate the cumulative probability (area to the left)
# This is for one tail test. For two tail test, we need to multiply cumulative_probability by a factor of 2 for both the tails
cumulative_probability = stats.norm.cdf(Z_Score)

# Convert the probability to a percentile (as a percentage)
percentile = cumulative_probability * 100

```

```python
print("Conversion Rate of Test", Conv_rate_A)
print("Conversion Rate of Control", Conv_rate_B)
print("Overall Lift", str(round((Conv_rate_B/Conv_rate_A - 1),2) * 100) + "%")
print("Z_Score", Z_Score)
print("P_Value", cumulative_probability)

if percentile > (1 - confidence_level):
    print("Control is significantly better")
else:
    print("Null hypothesis cant be rejected")
```
